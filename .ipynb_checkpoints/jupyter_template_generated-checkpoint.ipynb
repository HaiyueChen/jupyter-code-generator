```python
blob_names = ["folderfolder/mock/MOCK_DATA.csv","folderfolder/mock/mockmock/iis express.txt","folderfolder/mock/mockmock/output.txt","folderfolder/mock/mockmock/output.txt"]
```

```python
blob_service = BlockBlobService(account_name=account_name, sas_token=sas_token)
```

```python
#parse filenames so that it is only the filename not the path
#And use azure.storage.blob package to download them
file_names = []
for blob_name in blob_names:
    file_name = blob_name.split("/")[-1]
    file_names.append(file_name)
    blob_service.get_blob_to_path(container_name, blob_name, file_name)
```

```python
file_pointers_map = {}
for file_name in file_names:
    file_opened = open(file_name)
    file_pointers_map[file_name] = file_opened
```

File pointers can be accessed by using their name to index in the map

```python
csv_file = file_pointers_map[file_names[0]]
for line in csv_file:
    print(line)
```

```python
#Close the file pointers
for file_name in file_pointers_map:
    file_pointers_map[file_name].close()
```
